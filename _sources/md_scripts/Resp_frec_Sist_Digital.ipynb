# Respuesta en Frecuencia de Funciones de Transferencia digitales

#### Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

## Pasa Bajos - Orden 1

```python
# Defino el sistema
b = [0, 1] # numerador
a = [1, -0.5] # denominador
```

```python
# Calculo ceros y polos
zeros = np.roots(b)
print(zeros)
poles = np.roots(a)
print(poles)
```

Respuesta en Frecuencia

```python
# Figuras con dos subplots
fig, ax = plt.subplots(1,2)
fig.suptitle('Pasa Bajos - Orden 1', fontsize=22)
fig.set_size_inches(20, 10)

# grafico circulo unitario
theta = np.linspace(-np.pi,np.pi,201)
ax[0].plot(np.sin(theta), np.cos(theta), color='gray', linewidth=0.2)

# grafico el eje x-y
ax[0].axhline(y=0, color='gray', linewidth=1)
ax[0].axvline(x=0, color='gray', linewidth=1)

# Grafico polos y ceros
ax[0].plot(np.real(poles), np.imag(poles), 'Xb', label = 'Poles')

# Anotaciones
ax[0].annotate('poles',
            xy=(np.real(poles[0]), np.imag(poles[0])),
            xycoords='data',
            xytext=(30,60),
            textcoords='offset points',
            va='top',
            ha='left',
            bbox=dict(facecolor='w',
                      edgecolor='none',
                      boxstyle='round, pad=0'
                      ),
            arrowprops=dict(facecolor='b',
                            edgecolor='b',
                            width=2,
                            linewidth=0,
                            shrink=0.2,
                            headwidth=6
                            #headlength=8
                            )
           )

# create w array which travels around unit circle
w = np.linspace(-np.pi, np.pi, 201)
z = np.exp(1j*w)

# evalue H over the w array
H = np.polyval(b, z) / np.polyval(a, z)

# grafico el eje x-y
ax[1].axvline(x=0, color='gray', linewidth=1)

# plot the magnitude of H vs frequency.  H is a complex number array.
ax[1].plot(w, np.absolute(H))

# add lables
ax[0].set_title("Diagrama de Polos y Ceros", fontsize=20)
ax[0].set_xlabel("Eje Real", fontsize=16)
ax[0].set_ylabel("Eje Imaginario", fontsize=16)
ax[1].set_title("Respuesta en Frecuencia de H(z)", fontsize=20)
ax[1].set_ylabel("Magnitud |H(\u03A9)|", fontsize=16)
ax[1].set_xlabel("Frecuencia \u03A9", fontsize=16)
```

## Pasa Bajos - Orden 3

```python
# Defino el sistema
b = np.convolve(np.convolve([1., 1.],[1., 1.]),[1., 1.]) # numerador
a = np.convolve(np.convolve([1, -0.6],[1, -0.63+1j*0.5]),[1,-0.63-1j*0.5]) # denominador
```

```python
# Calculo ceros y polos
zeros = np.roots(b)
print(zeros)
poles = np.roots(a)
print(poles)
```

```python
# Figuras con dos subplots
fig, ax = plt.subplots(1,2)
fig.suptitle('Pasa Bajos - Orden 3', fontsize=22)
fig.set_size_inches(20, 10)

# grafico circulo unitario
theta = np.linspace(-np.pi,np.pi,201)
ax[0].plot(np.sin(theta), np.cos(theta), color='gray', linewidth=0.2)

# grafico el eje x-y
ax[0].axhline(y=0, color='gray', linewidth=1)
ax[0].axvline(x=0, color='gray', linewidth=1)

# Grafico polos y ceros
ax[0].plot(np.real(poles), np.imag(poles), 'Xb', label = 'Poles')
ax[0].plot(np.real(zeros), np.imag(zeros), 'or', label = 'Zeros')

# Anotaciones
ax[0].annotate('poles',
            xy=(np.real(poles[0]), np.imag(poles[0])),
            xycoords='data',
            xytext=(30,60),
            textcoords='offset points',
            va='top',
            ha='left',
            bbox=dict(facecolor='w',
                      edgecolor='none',
                      boxstyle='round, pad=0'
                      ),
            arrowprops=dict(facecolor='b',
                            edgecolor='b',
                            width=2,
                            linewidth=0,
                            shrink=0.2,
                            headwidth=6
                            #headlength=8
                            )
           )

ax[0].annotate('zeros',
            xy=(np.real(zeros[0]), np.imag(zeros[0])),
            xycoords='data',
            xytext=(-40,-40),
            textcoords='offset points',
            va='top',
            ha='right',
            bbox=dict(facecolor='w',
                      edgecolor='none',
                      boxstyle='round, pad=0'
                      ),
            arrowprops=dict(facecolor='r',
                            edgecolor='r',
                            width=2,
                            linewidth=0,
                            shrink=0.2,
                            headwidth=6
                            #headlength=8
                            )
           )

# create w array which travels around unit circle
w = np.linspace(-np.pi, np.pi, 201)
z = np.exp(1j*w)

# evalue H over the w array
H = np.polyval(b, z) / np.polyval(a, z)

# grafico el eje x-y
ax[1].axvline(x=0, color='gray', linewidth=1)

# plot the magnitude of H vs frequency.  H is a complex number array.
ax[1].plot(w, np.absolute(H))

# add lables
ax[0].set_title("Diagrama de Polos y Ceros", fontsize=20)
ax[0].set_xlabel("Eje Real", fontsize=16)
ax[0].set_ylabel("Eje Imaginario", fontsize=16)
ax[1].set_title("Respuesta en Frecuencia de H(z)", fontsize=20)
ax[1].set_ylabel("Magnitud |H(\u03A9)|", fontsize=16)
ax[1].set_xlabel("Frecuencia \u03A9", fontsize=16)
#plt.show()
```
