# Respuesta en Frecuencia de Sistema FIR

#### Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

### Sistema FIR Simetrico

```python
w = np.linspace(-np.pi, np.pi, 201)
exp_complx = np.exp(1j*w)

# Respuesta al impulso siemptrica
# h = [1, -1, 2, -1, 1]

H = exp_complx**(-2)*(2*1*np.cos(2*w) + 2*(-1)*np.cos(w) + 2)

Mag = np.abs(H)
Fase = np.angle(H)
```

Respuesta en Frecuencia

```python
# Figuras con dos subplots
fig, ax = plt.subplots(1,2)
fig.suptitle('Sistema FIR Simetrico', fontsize=22)
fig.set_size_inches(20, 10)

# grafico el eje x-y
ax[0].axvline(x=0, color='gray', linewidth=1)

# plot the magnitude of H vs frequency.  H is a complex number array.
ax[0].plot(w, Mag)

ax[1].plot(w, Fase)

# add lables
ax[0].set_title("Respuesta en Frecuencia: Magnitud", fontsize=20)
ax[0].set_xlabel("Frecuencia \u03A9", fontsize=16)
ax[0].set_ylabel("Magnitud |H(\u03A9)|", fontsize=16)
ax[1].set_title("Respuesta en Frecuencia: Fase", fontsize=20)
ax[1].set_ylabel("Fase H(\u03A9)", fontsize=16)
ax[1].set_xlabel("Frecuencia \u03A9", fontsize=16)
```
